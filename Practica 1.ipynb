**• Bubble Sort (Ordenamiento de Burbuja):** Compara pares de elementos adyacentes y los intercambia si están en el orden incorrecto. Se repite hasta que no se realicen intercambios.


```python
import random
import numpy as np
import timeit
import matplotlib.pyplot as plt

def bubble_sort(arr):
    n = len(arr)
    comparisons = 0
    passes = 0

    for i in range(n):
        swapped = False
        for j in range(0, n - i - 1):
            comparisons += 1
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                swapped = True
        passes += 1

        if not swapped:
            break

    return comparisons, passes

# Medir el tiempo de ejecución y el número de comparaciones
def measure_performance(size):
    arr = np.random.randint(1, 1000, size=size)
    execution_time = timeit.timeit(lambda: bubble_sort(arr.copy()), number=1)
    comparisons, passes = bubble_sort(arr.copy())
    return execution_time, comparisons, passes

# Parámetros
sizes = [100, 200, 300, 400, 500]
execution_times = []
comparisons_list = []
passes_list = []

# Medir el rendimiento para diferentes tamaños de datos
for size in sizes:
    execution_time, comparisons, passes = measure_performance(size)
    execution_times.append(execution_time)
    comparisons_list.append(comparisons)
    passes_list.append(passes)

# Graficar los resultados
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.plot(sizes, execution_times, marker='o')
plt.title('Tiempo de Ejecución vs Tamaño del Vector')
plt.xlabel('Tamaño del Vector')
plt.ylabel('Tiempo de Ejecución (segundos)')

plt.subplot(1, 2, 2)
plt.plot(sizes, comparisons_list, marker='o', label='Comparaciones')
plt.plot(sizes, passes_list, marker='o', label='Pasadas')
plt.title('Comparaciones, Pasadas vs Tamaño del Vector')
plt.xlabel('Tamaño del Vector')
plt.ylabel('Cantidad')
plt.legend()

plt.tight_layout()
plt.show()
```

**• Insertion Sort (Ordenamiento por Inserción):** Construye una lista ordenada uno a uno, insertando cada elemento en su posición correcta en la lista ya ordenada


```python
import random
import time
import matplotlib.pyplot as plt

def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and key < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key

def generate_sorted_array(size):
    return [i for i in range(1, size + 1)]

def measure_time(func, *args):
    start_time = time.time()
    func(*args)
    end_time = time.time()
    return end_time - start_time

sizes = [100, 200, 300, 400, 500]
time_results = []

for size in sizes:
    sorted_array = generate_sorted_array(size)
    time_taken = measure_time(insertion_sort, sorted_array.copy())
    time_results.append(time_taken)

plt.plot(sizes, time_results, marker='o')
plt.xlabel("Tamaño de la matriz")
plt.ylabel("Tiempo empleado (segundos)")
plt.title("Sorted Arrays")
plt.grid(True)
plt.show()
```

**• Selection Sort (Ordenamiento por Selección):** Encuentra el elemento mínimo de la lista y lo coloca en la posición correcta. Luego, repite el proceso para el resto de los elementos.

```python
import random
import time
import matplotlib.pyplot as plt

def selection_sort(arr):
    n = len(arr)
    for i in range(n - 1):
        min_index = i
        for j in range(i + 1, n):
            if arr[j] < arr[min_index]:
                min_index = j
        if min_index != i:
            arr[i], arr[min_index] = arr[min_index], arr[i]

# Generar un vector random de enteros
def generate_random_array(size):
    return [random.randint(1, 1000) for _ in range(size)]

# Medir el tiempo de ejecución del algoritmo de ordenamiento
def measure_time(func, *args):
    start_time = time.time()
    func(*args)
    end_time = time.time()
    return end_time - start_time

# Probar el algoritmo de ordenamiento y graficar los resultados
array_sizes = [10, 50, 100, 200, 500, 1000]
execution_times = []

for size in array_sizes:
    random_array = generate_random_array(size)
    time_taken = measure_time(selection_sort, random_array.copy())
    execution_times.append(time_taken)

plt.plot(array_sizes, execution_times, marker='o')
plt.xlabel("Tamaño de la matriz")
plt.ylabel("Tiempo empleado (segundos)")
plt.title("Selection Sort")
plt.grid()
plt.show()
```

**• Merge Sort (Ordenamiento por Mezcla):** Divide la lista en mitades más pequeñas, las ordena por separado y luego las combina en una lista ordenada más grande.

```python
import random
import time
import matplotlib.pyplot as plt

def merge_sort(arr):
    if len(arr) > 1:
        mid = len(arr) // 2
        left_half = arr[:mid]
        right_half = arr[mid:]

        merge_sort(left_half)
        merge_sort(right_half)

        i, j, k = 0, 0, 0

        while i < len(left_half) and j < len(right_half):
            if left_half[i] < right_half[j]:
                arr[k] = left_half[i]
                i += 1
            else:
                arr[k] = right_half[j]
                j += 1
            k += 1

        while i < len(left_half):
            arr[k] = left_half[i]
            i += 1
            k += 1

        while j < len(right_half):
            arr[k] = right_half[j]
            j += 1
            k += 1

def generate_random_array(size):
    return [random.randint(1, 1000) for _ in range(size)]

def measure_time(func, *args):
    start_time = time.time()
    func(*args)
    end_time = time.time()
    return end_time - start_time

sizes = [100, 200, 300, 400, 500]
time_results = []

for size in sizes:
    random_array = generate_random_array(size)
    time_taken = measure_time(merge_sort, random_array.copy())
    time_results.append(time_taken)

plt.plot(sizes, time_results, marker='o')
plt.xlabel("Tamaño de la matriz")
plt.ylabel("Tiempo empleado (segundos)")
plt.title("Merge Sort")
plt.grid(True)
plt.show()
```

**• Quick Sort (Ordenamiento Rápido):**Selecciona un elemento llamado "pivote" y reorganiza la lista de manera que los elementos más pequeños estén a su izquierda y los mayores a su derecha. Luego, se aplica recursivamente a las sublistas.

```python
import random
import time
import matplotlib.pyplot as plt

def quick_sort(arr):
    if len(arr) <= 1:
        return arr
    else:
        pivot = arr[random.randint(0, len(arr) - 1)]
        less_than_pivot = [x for x in arr if x < pivot]
        equal_to_pivot = [x for x in arr if x == pivot]
        greater_than_pivot = [x for x in arr if x > pivot]

        return quick_sort(less_than_pivot) + equal_to_pivot + quick_sort(greater_than_pivot)

def generate_sorted_array(size):
    return [i for i in range(1, size + 1)]

def measure_time(func, *args):
    start_time = time.time()
    func(*args)
    end_time = time.time()
    return end_time - start_time

sizes = [100, 200, 300, 400, 500]
time_results = []

for size in sizes:
    sorted_array = generate_sorted_array(size)
    time_taken = measure_time(quick_sort, sorted_array.copy())
    time_results.append(time_taken)

plt.plot(sizes, time_results, marker='o')
plt.xlabel("Tamaño de la matriz")
plt.ylabel("Tiempo empleado (segundos)")
plt.title("Quick Sort")
plt.grid(True)
plt.show()
```

**• Heap Sort (Ordenamiento por Montículos):** Convierte la lista en un montículo (estructura de datos especial), luego extrae repetidamente el elemento máximo (raíz del montículo) y lo coloca al final.

```python
import random
import time
import matplotlib.pyplot as plt

def heapify(arr, n, i):
    largest = i
    left = 2 * i + 1
    right = 2 * i + 2

    if left < n and arr[left] > arr[largest]:
        largest = left

    if right < n and arr[right] > arr[largest]:
        largest = right

    if largest != i:
        arr[i], arr[largest] = arr[largest], arr[i]
        heapify(arr, n, largest)

def heap_sort(arr):
    n = len(arr)

    for i in range(n // 2 - 1, -1, -1):
        heapify(arr, n, i)

    for i in range(n - 1, 0, -1):
        arr[i], arr[0] = arr[0], arr[i]
        heapify(arr, i, 0)

def generate_sorted_array(size):
    return [i for i in range(1, size + 1)]

def measure_time(func, *args):
    start_time = time.time()
    func(*args)
    end_time = time.time()
    return end_time - start_time

sizes = [100, 200, 300, 400, 500]
time_results = []

for size in sizes:
    sorted_array = generate_sorted_array(size)
    time_taken = measure_time(heap_sort, sorted_array.copy())
    time_results.append(time_taken)

plt.plot(sizes, time_results, marker='o')
plt.xlabel("Tamaño de la matriz")
plt.ylabel("Tiempo empleado (segundos)")
plt.title("Heap Sort")
plt.grid(True)
plt.show()
```

```python
import random
import time
import matplotlib.pyplot as plt

def bubble_sort(arr):
    # Implementación del algoritmo Bubble Sort
    iterations = 0
    n = len(arr)
    for i in range(n):
        for j in range(0, n - i - 1):
            iterations += 1
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    return iterations

def insertion_sort(arr):
    # Implementación del algoritmo Insertion Sort
    iterations = 0
    n = len(arr)
    for i in range(1, n):
        key = arr[i]
        j = i - 1
        while j >= 0 and key < arr[j]:
            iterations += 1
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key
    return iterations

def selection_sort(arr):
    # Implementación del algoritmo Selection Sort
    iterations = 0
    n = len(arr)
    for i in range(n):
        min_index = i
        for j in range(i + 1, n):
            iterations += 1
            if arr[j] < arr[min_index]:
                min_index = j
        arr[i], arr[min_index] = arr[min_index], arr[i]
    return iterations

def merge_sort(arr):
    # Implementación del algoritmo Merge Sort
    def merge(arr, left, mid, right):
        iterations = 0
        n1 = mid - left + 1
        n2 = right - mid
        L = arr[left:mid + 1]
        R = arr[mid + 1:right + 1]
        i = j = 0
        k = left
        while i < n1 and j < n2:
            iterations += 1
            if L[i] <= R[j]:
                arr[k] = L[i]
                i += 1
            else:
                arr[k] = R[j]
                j += 1
            k += 1
        while i < n1:
            arr[k] = L[i]
            i += 1
            k += 1
        while j < n2:
            arr[k] = R[j]
            j += 1
            k += 1
        return iterations

    def merge_sort_recursive(arr, left, right):
        iterations = 0
        if left < right:
            mid = (left + right) // 2
            iterations += merge_sort_recursive(arr, left, mid)
            iterations += merge_sort_recursive(arr, mid + 1, right)
            iterations += merge(arr, left, mid, right)
        return iterations

    n = len(arr)
    return merge_sort_recursive(arr, 0, n - 1)

def quick_sort(arr):
    # Implementación del algoritmo Quick Sort
    def partition(arr, low, high):
        pivot = arr[high]
        i = low - 1
        iterations = 0
        for j in range(low, high):
            iterations += 1
            if arr[j] < pivot:
                i += 1
                arr[i], arr[j] = arr[j], arr[i]
        arr[i + 1], arr[high] = arr[high], arr[i + 1]
        return i + 1, iterations

    def quick_sort_recursive(arr, low, high):
        iterations = 0
        if low < high:
            pi, iterations_partition = partition(arr, low, high)
            iterations += iterations_partition
            iterations += quick_sort_recursive(arr, low, pi - 1)
            iterations += quick_sort_recursive(arr, pi + 1, high)
        return iterations

    n = len(arr)
    return quick_sort_recursive(arr, 0, n - 1)

def heapify(arr, n, i,iterations=0):
    # Implementación de la función heapify para Heap Sort
    largest = i
    left = 2 * i + 1
    right = 2 * i + 2
    if left < n and arr[i] < arr[left]:
        largest = left
    if right < n and arr[largest] < arr[right]:
        largest = right
    if largest != i:
        arr[i], arr[largest] = arr[largest], arr[i]
        iterations += 1
        iterations += heapify(arr, n, largest,iterations)
    return iterations

def heap_sort(arr):
    # Implementación del algoritmo Heap Sort
    def build_heap(arr, iterations=0):
        n = len(arr)
        for i in range(n // 2 - 1, -1, -1):
            iterations += heapify(arr, n, i)
        return iterations

    iterations = build_heap(arr)
    n = len(arr)
    for i in range(n - 1, 0, -1):
        arr[i], arr[0] = arr[0], arr[i]
        iterations += heapify(arr, i, 0)
    return iterations

# Comparar los tiempos de ejecución y el número de iteraciones
algorithms = {
    "Bubble Sort": bubble_sort,
    "Insertion Sort": insertion_sort,
    "Selection Sort": selection_sort,
    "Merge Sort": merge_sort,
    "Quick Sort": quick_sort,
    "Heap Sort": heap_sort
}

n_values = [100, 500, 1000, 2000, 3000]
iterations_data = {name: [] for name in algorithms}
time_data = {name: [] for name in algorithms}

# Valores por defecto para las pruebas aleatorias
min_value = 1
max_value = 10000

for n in n_values:
    random_vector = [random.randint(min_value, max_value) for _ in range(n)]
    for name, algorithm in algorithms.items():
        vector_copy = random_vector.copy()
        start_time = time.time()
        iterations = algorithm(vector_copy)
        end_time = time.time()
        elapsed_time = end_time - start_time
        iterations_data[name].append(iterations)
        time_data[name].append(elapsed_time)

# Crear gráficos
plt.figure(figsize=(12, 6))

# Gráfico de número de iteraciones
plt.subplot(1, 2, 1)
for name, iterations in iterations_data.items():
    plt.plot(n_values, iterations, marker='o', label=name)
plt.xlabel('Tamaño del Conjunto de Datos (n)')
plt.ylabel('Número de Iteraciones')
plt.title('Número de Iteraciones vs Tamaño del Conjunto de Datos')
plt.legend()

# Gráfico de tiempo de ejecución
plt.subplot(1, 2, 2)
for name, times in time_data.items():
    plt.plot(n_values, times, marker='o', label=name)
plt.xlabel('Tamaño del Conjunto de Datos (n)')
plt.ylabel('Tiempo de Ejecución (segundos)')
plt.title('Tiempo de Ejecución vs Tamaño del Conjunto de Datos')
plt.legend()

plt.tight_layout()
plt.show()
```
